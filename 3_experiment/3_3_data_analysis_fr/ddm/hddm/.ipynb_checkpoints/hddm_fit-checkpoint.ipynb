# SETUP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import copy

%matplotlib inline
```

```python
import hddm
```

```python
# Load data from csv file into a NumPy structured array
data = hddm.load_csv('data/foodreg_fmri.csv')
```

```python
data.head()
```

```python
data = data.drop(["Unnamed: 0"], axis=1)
data.head()
```

```python
data = hddm.utils.flip_errors(data)

fig = plt.figure()
ax = fig.add_subplot(111, xlabel='RT', ylabel='count', title='RT distributions')
for i, subj_data in data.groupby('subj_idx'):
    subj_data.rt.hist(bins=20, histtype='step', ax=ax)
```

# STATS

## Calculate correlations 

Look at pre-liking vs. response preference

```python
corrs = data.groupby(['cond', 'subj_idx'])[['resp','pre_liking']].corr()
corrs = corrs.iloc[::2]
corrs = corrs.reset_index()
corrs = corrs.drop(['level_2', 'resp'], axis=1)

x = corrs.groupby('cond')['pre_liking'].describe()
x['se'] = x['std']/np.sqrt(x['count'])
x
```

## Plot

```python
plt.hist(corrs['pre_liking'][corrs['cond']=='Decrease'], bins=15)
plt.hist(corrs['pre_liking'][corrs['cond']=='Health'], bins = 15, alpha = 0.9)
plt.hist(corrs['pre_liking'][corrs['cond']=='Natural'], bins =15, alpha = 0.9)
```

```python
def calc_quants(ppc, quantiles, output):
    
    ppc = ppc.reset_index(level=['node', 'sample'])
    
    # create sampled response column
    ppc['resp_sampled'] = [0 if i < 0 else 1 for i in ppc.rt_sampled]
    
    # make all rts positive
    ppc.rt_sampled = [np.abs(i) for i in ppc.rt_sampled]
    ppc.rt = [np.abs(i) for i in ppc.rt]
    
    # create quantiles by subject for rt
    actual = ppc.groupby(['node', 'response']).rt.quantile(quantiles)
    sim = ppc.groupby(['node', 'sample', 'response']).rt_sampled.quantile(quantiles)
    
    actual = actual.reset_index() # get rid of multi-index
    sim = sim.reset_index()

    actual.rename(index=str, columns={"level_2": "quantile"}, inplace=True)
    sim.rename(index=str, columns={"level_3": "quantile"}, inplace=True)
    
    sim_mean = sim.groupby(['node', 'quantile', 'response']).rt_sampled.mean().reset_index()
    sim_sem = sim.groupby(['node', 'quantile', 'response']).rt_sampled.sem().reset_index()
    sim_sem.rename(index=str, columns={"rt_sampled": "sem"}, inplace=True)

    # Match indexes
    actual.reset_index(drop=True, inplace=True) 
    sim_sem.reset_index(drop=True, inplace=True) 

    rt_sub_quantiles = pd.concat([actual, sim_mean.iloc[:,3], sim_sem.iloc[:,3]], axis=1)
        
    a = rt_sub_quantiles.groupby(['quantile', 'response']).rt.mean() 
    b = rt_sub_quantiles.groupby(['quantile', 'response']).rt_sampled.mean() 
    c = rt_sub_quantiles.groupby(['quantile', 'response']).rt_sampled.sem().rename('sem')
    rt_group_quantiles = pd.concat([a,b,c], axis = 1).reset_index()
    
    if output == 'subject':
        output_df = rt_sub_quantiles
    elif output == 'group':
        output_df = rt_group_quantiles

    # make version column yes/no & actual/sim
    actual = copy.copy(output_df)
    sim = copy.copy(output_df)
    actual["version"] = 'actual: yes'
    actual["version"][actual.response==0] = 'actual: no'
    sim["version"] = 'sim: yes'
    sim["version"][sim.response==0] = 'sim: no'
    
    # rename and concat into one df
    sim.drop(columns='rt', axis=1, inplace=True)
    sim.rename(index=str, columns={"rt_sampled": "rt"}, inplace=True)
    actual.drop(columns='rt_sampled', axis=1, inplace=True)
    rt_q_plot = pd.concat([actual, sim])
    
    return(rt_q_plot)
```

```python
def plot_quants(df, style):
    
    if style == 'a':
        plt.errorbar(x = df['quantile'][df.version=='actual: no'],
                     y = df['rt'][df.version=='actual: no'],
                     yerr=0,
                    label = "Actual: No")
        plt.errorbar(x = df['quantile'][df.version=='actual: yes'],
                     y = df['rt'][df.version=='actual: yes'],
                     yerr=0,
                    label = "Actual: Yes")
        plt.errorbar(x = df['quantile'][df.version=='sim: no'],
                     y = df['rt'][df.version=='sim: no'],
                     yerr=df['sem'][df.version=='sim: no'],
                    label = "Sim: No")
        plt.errorbar(x = df['quantile'][df.version=='sim: yes'],
                     y = df['rt'][df.version=='sim: yes'],
                     yerr=df['sem'][df.version=='sim: yes'],
                    label = "Sim: Yes")
        plt.xlim(0,1)
        plt.legend(loc=2)
        
    if style == 'b':
        plt.errorbar(x = np.linspace(0,3,100),
                     y = np.linspace(0,3,100),
                     yerr=0,
                     linestyle="--",
                    label = "Actual")
        plt.errorbar(x = df['rt'][df.version=='actual: yes'],
                     y = df['rt'][df.version=='sim: yes'],
                     yerr=df['sem'][df.version=='sim: yes'],
                     linestyle="", marker="o", markersize=5,
                    label = "Sim: Yes")
        plt.errorbar(x = df['rt'][df.version=='actual: no'],
                     y = df['rt'][df.version=='sim: no'],
                     yerr=df['sem'][df.version=='sim: no'],
                     linestyle="", marker="o", markersize=5,
                    label = "Sim: No")
        plt.xlim(np.min([df['rt'][df.version=='sim: no'],
                         df['rt'][df.version=='sim: yes']]),
                 np.max([df['rt'][df.version=='sim: no'],
                         df['rt'][df.version=='sim: yes']]))
        plt.legend(loc=2)



        
```

```python
import pymc as pm
import numpy as np
import pymc.progressbar as pbar

def _parents_to_random_posterior_sample(bottom_node, pos=None):
    """Walks through parents and sets them to pos sample."""
    for i, parent in enumerate(bottom_node.extended_parents):
        if not isinstance(parent, pm.Node): # Skip non-stochastic nodes
            continue

        if pos is None:
            # Set to random posterior position
            pos = np.random.randint(0, len(parent.trace()))

        assert len(parent.trace()) >= pos, "pos larger than posterior sample size"
        parent.value = parent.trace()[pos]

def _post_pred_generate(bottom_node, samples=500, data=None, append_data=True):
    """Generate posterior predictive data from a single observed node."""
    datasets = []
    ##############################
    # Sample and generate stats
    for sample in range(samples):
        _parents_to_random_posterior_sample(bottom_node)
        # Generate data from bottom node
        sampled_data = bottom_node.random()
        if append_data and data is not None:
            sampled_data.reset_index(inplace=True)  # Only modification of original Kabuki code
            sampled_data = sampled_data.join(data.reset_index(), lsuffix='_sampled')
        datasets.append(sampled_data)
    return datasets

def post_pred_gen(model, groupby=None, samples=500, append_data=False, progress_bar=True):
    results = {}

    # Progress bar
    if progress_bar:
        n_iter = len(model.get_observeds())
        bar = pbar.progress_bar(n_iter)
        bar_iter = 0
    else:
        print("Sampling...")

    if groupby is None:
        iter_data = ((name, model.data.ix[obs['node'].value.index]) for name, obs in model.iter_observeds())
    else:
        iter_data = model.data.groupby(groupby)

    for name, data in iter_data:
        node = model.get_data_nodes(data.index)

        if progress_bar:
            bar_iter += 1
            bar.update(bar_iter)

        if node is None or not hasattr(node, 'random'):
            continue # Skip

        ##############################
        # Sample and generate stats
        datasets = _post_pred_generate(node, samples=samples, data=data, append_data=append_data)
        results[name] = pd.concat(datasets, names=['sample'], keys=list(range(len(datasets))))

    if progress_bar:
        bar_iter += 1
        bar.update(bar_iter)

    return pd.concat(results, names=['node'])
```
